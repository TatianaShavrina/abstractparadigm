Напишите функцию **substitute_words(paradigm, lst)**, принимающую на вход абстрактную парадигму **paradigm** и список строк **lst**, и возвращающую результат подстановки этих строк в парадигму. Гарантируется, что все компоненты парадигмы содержат одни и те же переменные, а также что выполнены все требования из задания к функции **substitute**.

```python
## Напишите код
def substitute_words(paradigm, lst):
    par=paradigm.split("#")
    res=[]
    for elem in par:
        for j in lst:
            elem=re.sub(r'(\+)?\d(\+)?', j, elem, count=1)
        res.append(elem)
    return res
##
print(substitute_words('1+а#1+и#1+и#1#1+е#1+ам#1+у#1+и#1+ой#1+ами#1+е#1+ах', ['интриг']))
print(substitute_words('1+е+2+3+о#1+е+2+3+о#1+ё+2+3+а#1+е+2+3+а#1+ё+2+е+3#1+е+2+3+у\
#1+ё+2+3+ам#1+е+2+3+о#1+ё+2+3+а#1+е+2+3+ом#1+ё+2+3+ами#1+е+2+3+е#1+ё+2+3+ах',
                 ['рем', 'с',  'л']))
print(substitute_words('1+о+2#1+о+2#1+2+и#1+2+а#1+2+ов#1+2+у#1+2+ам#1+о+2#1+2+и#1+2+ом#\
1+2+ами#1+2+е#1+2+ах', ['нос', 'к']))
```

Для начала реализуйте функцию **all_partitions(word, n)**, возвращающую список всех возможных разбиений слова **word** на **n** непустых подстрок. При этом может оказаться полезной приведённая ниже функция **generate_monotone_sequences(min_value, max_value, n)**, возвращающая все монотонно возрастающие последовательности длины **n** из элементов в полуинтервале [**min_value**, **max_value**). Замечание: **n** в разных функциях могут принимать различные значения.

```python
def next_monotone_sequence(seq, max_value):
    """
    Порождает следующую монотонную последовательность
    со значениями меньше max_value
    """
    if seq[-1] < max_value - 1:
        return seq[:-1] + [seq[-1] + 1]
    n = len(seq)
    for i in range(len(seq) - 2, -1, -1):
        if seq[i] + 1 < seq[i+1]:
            # увеличиваем i-ый элемент на 1 и заполняем хвост
            # минимально возможными значениями
            # (каждое следующее на 1 больше предыдущего)
            return seq[:i] + list(range(seq[i] + 1, seq[i] + (n - i + 1)))
    return None

def generate_monotone_sequences(min_value, max_value, n):
    """
    Порождает все монотонные последовательности длины n 
    из элементов в полуинтервале [min_value; max_value)
    """
    if min_value + n > max_value:
        return []
    if n == 0:
        return [[]]  # пустая монотонная последовательность
    # будем перебирать монотонные последовательности
    # в лексикографическом порядке
    last_seq = list(range(min_value, min_value + n))
    answer = []
    while last_seq is not None:
        answer.append(last_seq)
        last_seq = next_monotone_sequence(last_seq, max_value)
    return answer

print(generate_monotone_sequences(0, 3, 1)) # [[0], [1], [2]]
print(generate_monotone_sequences(0, 3, 2)) # [[0, 1], [0, 2], [1, 2]]
print(generate_monotone_sequences(2, 6, 4)) # [[2, 3, 4, 5]]
print(generate_monotone_sequences(2, 7, 4)) # [[2, 3, 4, 5], [2, 3, 4, 6], [2, 3, 5, 6], [2, 4, 5, 6], [3, 4, 5, 6]]
print(generate_monotone_sequences(0, 3, 4)) # []


## Напишите код
def all_partitions(word, n):
    a=n-1
    min=0
    max=len(word)-1
    indexes=generate_monotone_sequences(min, max, a)
#    print (indexes)
    spisokspiskov=[]
    for i in indexes:
        z=[]
        left=0
        for j in i:
            f=word[left:j+1]
            left=j+1
            z.append(f)
        z.append(word[left:len(word)])    
        spisokspiskov.append(z)
#    print (spisokspiskov)
    return spisokspiskov
print(all_partitions('лес', 2)) # [['л', 'ес'], ['ле', 'с']]
print(all_partitions('песок', 3)) 
# [['п', 'е', 'сок'], ['п', 'ес', 'ок'], ['п', 'есо', 'к'], ['пе', 'с', 'ок'], ['пе', 'со', 'к'], ['пес', 'о', 'к']]
print(all_partitions('песок', 1)) # [['песок']]
print(all_partitions('песок', 5)) # [['п', 'е', 'с', 'о', 'к']]
```

Реализуйте функцию **preprocess_pattern(pattern)**, получающую на вход шаблон **pattern** и возвращающую две величины: список его постоянных фрагментов **const_fragments_list** и список **consecutive_variable_fragments_numbers**, имеющий длину $k+1$, где $k$ --- число постоянных фрагментов и содержащий в своём $i$-ом элементе число переменных между $i$-ым $(i+1)$-ым слева постоянным фрагментом шаблона. При этом **consecutive_variable_fragments_numbers**[0] равно числу переменных слева от первого постоянного фрагмента, а **consecutive_variable_fragments_numbers**[k] --- справа от последнего.

```python
## Напишите код
def preprocess_pattern(pattern):
    s=pattern.split('+')
    const_fragments_list=[]
    consecutive_variable_fragments_numbers=[]
    for i in range(len(s)):
#        if s[i].isdigit()==True:
#            consecutive_variable_fragments_numbers.append(i)
        if s[i].isalpha()==True:
            const_fragments_list.append(s[i])
    sb=re.sub('(\+)', '', pattern, count=0)
    spisok_podstrok=re.findall('\d\d?\D\D?', sb)
    for i in spisok_podstrok:
        sum=0
        ir=list(i)
        for j in ir:
            if j.isdigit()==True:
                sum=sum+1
        consecutive_variable_fragments_numbers.append(sum)
        
    if s[len(s)-1].isalpha()==True:
        consecutive_variable_fragments_numbers.append(0)
    else:
        consecutive_variable_fragments_numbers.append(1)   
    return const_fragments_list, consecutive_variable_fragments_numbers

    
    
##

print(preprocess_pattern('1+ин')) # ['ин'], [1, 0]
print(preprocess_pattern('1+о+2'))  # ['о'], [1, 1]
print(preprocess_pattern('1+е+2+3+о'))  # ['е', 'о'], [1, 2, 0]
```

Реализуйте функцию **find_monotone(s, lst)**, которая получает на вход строку **s** и список строк **lst**=[$s_1$, $\ldots$, $s_m$] и находит все позиции подстрок $s_1$, $\ldots$, $s_m$ при условии, что подстрока $s_j$ начинается не левее конца строки $s_{j-1}$. Для этого сначала найдите для каждой строки $s_j$ все её позиции в $s$ (для этого можно воспользоваться функцией re.finditer и полями start и end у класса re.match или самостоятельно написать соответствующую функцию), получив списки позиций L_1, $\ldots$, L_m. Под позицией строки $t$ в $s$ в данном случае понимается пара индексов $(i, j)$, такая что $s[i:j] = t$ (например, подстрока 'аб' в строке 'абракадабра') имеет позиции (0, 2) и (7, 9).

После этого выделите все такие последовательности пар индексов, что первая пара лежит в $L_1$, вторая в $L_2$ и так далее, при условии, что каждая следующая пара начинается не левее конца предыдущей. Для этого сначала выделите все последовательности, чьи элементы принадлежат указанным спискам, с помощью функции product модуля itertools, а потом отберите из них возрастающие.

```python
## Напишите код
#тут вроде были исправленные ошибки в ответах, я посмотрела, но продолжила делать в старом ноутбуке
import itertools

def find_monotone(s, lst):
    posicii_podstrok=[]
    for elem in lst:
        c=[]
        for m in re.finditer(elem, s):
            a=[]
            a.append(m.start())
            a.append(m.end())
            b=tuple(a)
            c.append(b)
        posicii_podstrok.append(c)
    combin = itertools.product(*posicii_podstrok)
    combin = list(list(combination) for combination in combin)
    combin_final = []
    for elem in combin:
        if len(elem) > 1:
            for n in range(len(elem)-1):
                if elem[n][1] < elem[n+1][0]:
                    combin_final.append(elem)
        else:
            combin_final.append(elem)

    return combin_final

##
print(find_monotone('ремесло', ['е', 'о'])) # [[(2,3), (6, 7)], [(3,4), (6, 7)]]
print(find_monotone('желание', ['ан', 'е'])) # [[(3,5), (6, 7)]]
print(find_monotone('математика', ['ма', 'ат'])) # [[(0,2), (5, 6)]]
```

Реализуйте функцию **extract_uncovered(s, pairs)**, получающую строку **s** и список пар индексов **pairs**= [$(i_1, j_1)$, $\ldots$, $(i_k, j_k)$] и возвращающая список из $k+1$ подстрок $s$, остающиеся после удаления из неё подстрок $s[i_1: j_1], \ldots, s[i_k: j_k]$. При этом Вы можете считать, что удаляемые подстроки не пересекаются и каждая следующая идёт не левее конца предыдущей (то есть выполняется условие $j_l < i_{l+1}$ для всех $l$).

```python
## Напишите код
def extract_uncovered(s, pairs):
    stroka=s
    spisok=[]
    stroka_new=''
    for i in pairs:
        min=i[0]
        max=i[1]
        for j in range(min,max):
            spisok.append(j)
#    print (spisok)
    for s in range(len(stroka)):
        if s not in spisok:
            stroka_new=stroka_new+stroka[s]
        else:
            stroka_new=stroka_new+'+'
    
    f='+1+'
    stroka_new=re.sub('(\+){3}', f, stroka_new, count=0)
#    print(stroka_new)
    stroka_new=re.sub('(\+){2}', '+', stroka_new, count=0)
    spisok2=stroka_new.split('+')
    for i in range(len(spisok2)):
        if spisok2[i]=="1":
            spisok2[i]=''
#    print(stroka_new)
    return spisok2
##

print(extract_uncovered('ремесло', [(1,2), (6, 7)])) # ['р', 'есл']
print(extract_uncovered('желание', [(3,5), (6, 7)])) # ['жел', 'и', '']
print(extract_uncovered('математика', [(0,2), (5, 6)])) # ['', 'тем', 'ика']
print(extract_uncovered('корова', [(1,2), (2, 4)])) # ['к', '', 'ва']
```

Реализуйте функцию **extract_variables(pattern, word)**, находящую все возможные способы значения переменных при условии, что слово **word** подходит под шаблон **pattern**. Например, **extract_variables**('1+о+2', 'носок') = [['н', 'сок'], ['нос', 'к']]. Для этого воспользуйтесь написанными в предыдущих $3$ вопросах функциями. Вам нужно все возможные способы разбить слово на подстроки постоянными фрагментами шаблона **pattern**, а потом выделить в этих фрагментах нужное число переменных, которое можно извлечь из списка **consecutive_variable_fragments_numbers**, который вернёт функция **preprocess_pattern**. Таким образом Вы можете найти способы расставить переменные в каждом из участков между постоянными фрагментами.

Чтобы получить все способы расставить все переменные, к полученным спискам значений нужно применить функции product и chain из модуля itertools.

```python
## Напишите код
def extract_variables(pattern, word):
    [const_fragments_list, consecutive_variable_fragments_numbers] = preprocess_pattern(pattern)
    const_fragm_positions = find_monotone(word, const_fragments_list)
    itog = []
    g = len(consecutive_variable_fragments_numbers)
    for i in const_fragm_positions:
        fragment_replacements = []
        fragments = extract_uncovered(word, i)
        bool_criteria = True
        for k in range(g):
            n = consecutive_variable_fragments_numbers[k]
            m = fragments[k]
            if m == '' and n != 0:
                bool_criteria = False
            if m != '':
                if n == 0:
                    bool_criteria = False
                else:
                    fragments_2 = all_partitions(m, n)
                    if len(fragment_replacements) > 0:  
                        fragment_replacements = itertools.product(fragment_replacements, fragments_2)
                        fragment_replacements = list(fragment_replacements)
                        fragment_replacements = list(list(itertools.chain(*elem)) for elem in fragment_replacements)
                    else:
                        fragment_replacements = itertools.chain(fragment_replacements, fragments_2)
                        fragment_replacements = list(fragment_replacements)
        if bool_criteria == True:
            itog = itertools.chain(itog, fragment_replacements)
            itog = list(itog)
    return itog
##


print(extract_variables('1+а', 'палатка'))  # [['палатк']]
print(extract_variables('1+ь+2+а', 'свадьба'))  # [['свад', 'б']]
print(extract_variables('1+е+2+3+о', 'ремесло'))  # [['р', 'м', 'есл'], ['р', 'ме', 'сл'], ['р', 'мес', 'л'], ['рем', 'с', 'л']]
print(extract_variables('1+2+я', 'песня'))  # [['п', 'есн'], ['пе', 'сн'], ['пес', 'н']]
```
